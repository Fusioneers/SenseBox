```python
hallog = pd.read_csv(os.path.join(ABS_PATH, 'data', 'HALLOG00.CSV'))

timestamps = hallog.dropna()['timestamp']

delta = timestamps.diff()

delta = delta[20:]

delta = 1/delta

# Every value smaller than 0.05 is set to 0
delta = delta[delta > 0.05]

df = pd.DataFrame({'delta': delta/1000})


plt.plot(df)
plt.show()
df.describe()
```
